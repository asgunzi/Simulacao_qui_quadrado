```python
#Simulação para entender a distribuição Qui quadrado

import numpy as np
import matplotlib.pyplot as plt
```

```python
k = 1
N = 1000 #Numero de amostras

chi_squared_stat =[]
for _ in range(N):
    x = np.random.normal(loc=0, scale=1, size=k)
    x_squared = x**2
    chi_squared_stat.append(np.sum(x_squared))

plt.hist(chi_squared_stat, bins=30, density=True, alpha=0.6, color='g')
plt.title('Distribuição Qui-quadrado - Simulação')

plt.xlabel('Valor')
plt.ylabel('Densidade')
plt.show()

print(f'Média da distribuição Qui-quadrado: {np.mean(chi_squared_stat):.2f}')
```

```python
#Usando função pronta do numpy

graus_liberdade = 5
dist_chi = np.random.chisquare(df = graus_liberdade, size=1000)


plt.hist(dist_chi, bins=30, density=True, alpha=0.6, color='b')
plt.title('Distribuição Qui-quadrado - Função Pronta')
plt.xlabel('Valor')
plt.ylabel('Densidade')
plt.show()
```
